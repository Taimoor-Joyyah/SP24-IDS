```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('world_pop.csv')
df.head(5)
```

```python
top_countries = df.nlargest(10, 'year_2020')['country']
top_countries
```

```python
plt.figure(figsize=(10, 6))

for country in top_countries:
    country_data = df[df['country'] == country]
    country_data = country_data.drop(columns=['country'])
    plt.plot(country_data.columns, country_data.values[0], label=country)

plt.title('Top 10 Highest Populated Countries (1960-2020)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)

plt.xticks(df.columns[1::5], rotation=45)

plt.legend()
plt.show()
```
