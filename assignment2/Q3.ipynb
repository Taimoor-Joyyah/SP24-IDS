```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diamonds.csv')

df_filtered = df[(df['clarity'] == 'SI2') & (df['color'] == 'E')]
```

```python
color_map = {'Ideal': 'blue', 'Premium': 'green', 'Good': 'red', 'Very Good': 'purple', 'Fair': 'orange'}
colors = df_filtered['cut'].map(color_map)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['carat'], df_filtered['price'], c=colors, alpha=0.5)

plt.title('Relationship between Carat and Price of Diamonds', fontsize=14)
plt.xlabel('Carat', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.grid(True)
plt.legend(color_map.keys(), title='Cut', title_fontsize='13', fontsize='12')

plt.show()
```
