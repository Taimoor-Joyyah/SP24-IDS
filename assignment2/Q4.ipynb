```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('nuclear_waste_sites.csv')
```

```python
usa_map = folium.Map(location=[39.65, -97.95], zoom_start=4)

for index, row in df.iterrows():
    folium.Marker([row['lat'], row['lon']], popup=row['text']).add_to(usa_map)

usa_map
```
