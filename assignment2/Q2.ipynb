```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('world_pop.csv')
```

```python
least_populous_2015 = data[['country', 'year_2015']].sort_values(by='year_2015').head(10)
least_populous_2015.plot(x='country', y='year_2015', kind='bar', title='10 Least Populous Countries in 2015')
plt.show()
```

```python
countries = ['Pakistan', 'India', 'United States', 'United Kingdom']
population_change = data[data['country'].isin(countries)][['country', 'year_1970', 'year_2010']]
population_change['Change'] = (population_change['year_2010'] - population_change['year_1970']) / 10 ** 6
population_change[['country', 'Change']].plot(x='country', y='Change', kind='bar',
                                              title='Population Change from 1970 to 2010 (in millions)')
plt.show()
```

```python
pakistan_growth = data[data['country'] == 'Pakistan'].loc[:, 'year_2010':'year_2020'].T
pakistan_growth.columns = ['Population']
pakistan_growth['Growth'] = pakistan_growth['Population'].pct_change() * 100 
pakistan_growth['Growth'].plot(kind='line', title='Pakistan Population Growth from 2010 to 2020 (%)')
plt.show()
```
